```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image in color (BGR)
im = cv2.imread("C:\\Users\\Isabelle\\Desktop\\FilterImages\\images\\img4.jpg", cv2.IMREAD_COLOR)

# Initializing parameter setting using cv2.SimpleBlobDetector function
params = cv2.SimpleBlobDetector_Params()

# Filter by area (value for area here defines the pixel value)
params.filterByArea = True
params.minArea = 10

# Filter by circularity
params.filterByCircularity = True
params.minCircularity = 0.40

# Filter by convexity
params.filterByConvexity = True
params.minConvexity = 0.1

# Filter by inertia ratio
params.filterByInertia = True
params.minInertiaRatio = 0.01

# Creating a blob detector using the defined parameters
detector = cv2.SimpleBlobDetector_create(params)

# Detecting the blobs in the image
keypoints = detector.detect(im)

# Drawing the blobs that have been filtered with green on the image
blobs = cv2.drawKeypoints(im, keypoints, np.array([]), (0, 255, 0),
                          cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Setting the grid size
plt.figure(figsize=(12, 5))

# Displaying the original image (color)
plt.subplot(121)
plt.title('Original')
plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib

# Displaying the image with blobs detected
plt.subplot(122)
plt.title('Blobs\nArea: {}, Circularity: {}, Convexity: {}, Inertia: {}'.format(round(params.minArea, 1),
                                                                                 round(params.minCircularity, 1),
                                                                                 round(params.minConvexity, 1),
                                                                                 round(params.minInertiaRatio, 3)))
plt.imshow(cv2.cvtColor(blobs, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregue a imagem original
image = cv2.imread('C:/Users/Isabelle/Desktop/FilterImages/images/img4.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Converta a imagem para tons de cinza
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Define os parmetros para o Canny
low_threshold = 250
high_threshold = 350

# Aplique o detector de bordas Canny
edges = cv2.Canny(gray, low_threshold, high_threshold)

# Define os parmetros para a transformada de Hough
rho = 1.5
theta = np.pi/180
threshold = 50
min_line_length = 20
max_line_gap = 5

# Crie uma figura com 1 linha e 3 colunas
plt.figure(figsize=(15, 5))

# Subplot 1: Imagem Original
plt.subplot(131)
plt.title('Original Image')
plt.imshow(image)
plt.axis('off')

# Subplot 2: Deteco de Bordas
plt.subplot(132)
plt.title(f'Edge detection, low_threshold = {low_threshold}, high_threshold = {high_threshold}')
plt.imshow(edges, cmap='gray')
plt.axis('off')

# Subplot 3: Transformada de Hough
line_image = np.copy(image)
lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]),
                        min_line_length, max_line_gap)
for line in lines:
    for x1, y1, x2, y2 in line:
        cv2.line(line_image, (x1, y1), (x2, y2), (255, 0, 0), 2)

plt.subplot(133)
plt.title(f'Hough transform, rho = {rho}, theta = {round(theta,2)},\n threshold = {threshold}, min_line_length = {min_line_length}, max_line_gap = {max_line_gap}')
plt.imshow(line_image)
plt.axis('off')

# Ajuste o layout para evitar sobreposies
plt.tight_layout()

# Exiba os subplots
plt.show()
```
